# Douglass lab amp problem set

Problem:

Show quantitatively that our existing amplifier, the Stanford SR570 (manual attached), is too slow for 30 fps resonant scanning at 512 x 512 pixels.

OK, what are the relevant values?

* The time per pixel, because we need at least one sample per pixel, that should give us the max frequency that we need.

* The maximum frequency of the amplifiers

```python
fps = 30
pix_per_line = 512
total_pixels = pix_per_line * pix_per_line

spf = 1/fps
sec_per_pixel = spf / total_pixels

max_freq = 1/sec_per_pixel

print(f'The maximum frequency we need is {max_freq/1000000} MHz')
```

What is the max frequency of the amplifiers?

* 1 MHz

There's our problem!

What is the frequency we need with a range of resolutions and speeds?

```python
for fps in [1, 10, 20, 30, 40]:
    for pix_per_line in [128, 256, 512, 1024]:
        total_pixels = pix_per_line * pix_per_line

        spf = 1/fps
        sec_per_pixel = spf / total_pixels

        max_freq = 1/sec_per_pixel

        print(f'The maximum frequency we need at {pix_per_line} x {pix_per_line} and {fps} fps is {max_freq/1000000} MHz')
    print('')
    
```
